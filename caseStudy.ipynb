```python
import pandas as pd
import PyPDF2
import matplotlib.pyplot as plt
import seaborn as sns
import re
from tabulate import tabulate
customer_data = pd.read_csv('/content/dataset.csv')
print("Customer data successfully loaded.")

def extract_castes_from_pdf(pdf_path):
    caste_list = []
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            text = page.extract_text()
            cleaned_text = re.sub(r'\d+\.\s*', '', text)
            castes_on_page = [caste.strip().lower() for caste in cleaned_text.split(',')]
            caste_list.extend(castes_on_page)

    return list(set([caste for caste in caste_list if caste]))

# Extract caste names
caste_names = extract_castes_from_pdf('/content/LatestListofScheduledtribes.pdf')
print(f"{len(caste_names)} caste names extracted from the PDF.")
normalized_caste_names = [caste.lower() for caste in caste_names]

matched_names = []
matched_caste_names = []

def check_caste_match(row):
    names = [row['Full Name'], row['Father Name'], row['Mother Name'], row['Spouse Name']]
    normalized_names = [name.lower() for name in names if pd.notna(name)]

    matched_castes = []  # To store matched castes for this row
    for name in normalized_names:
        for caste in normalized_caste_names:
            if caste in name:
                matched_names.append(name)
                matched_castes.append(caste)
                matched_caste_names.append(caste)

    if matched_castes:
        return pd.Series([1, ', '.join(matched_castes)])
    else:
        return pd.Series([0, None])

customer_data[['Caste Match', 'Matched Castes']] = customer_data.apply(check_caste_match, axis=1)
print(f"Total caste matches found: {customer_data['Caste Match'].sum()}")

# Step 4: Count total matches and state-wise matches
total_matches = customer_data['Caste Match'].sum()
matches_by_state = customer_data[customer_data['Caste Match'] == 1].groupby('State').size()

# Step 5: State with maximum matches

max_matches_state = matches_by_state.idxmax()
max_matches_count = matches_by_state.max()
print(f"The state with the maximum number of matches is: {max_matches_state} with {max_matches_count} matches.")


# Step 6: Match percentage by state
total_customers_by_state = customer_data.groupby('State').size()
match_percentage_by_state = (matches_by_state / total_customers_by_state) * 100
print("Match percentage by state:\n", match_percentage_by_state.fillna(0))

# Step 7: Least and most frequent matched castes
matched_caste_counts = pd.Series(matched_caste_names).value_counts()

# Most frequent matched caste
most_frequent_matched_caste = matched_caste_counts.idxmax()
most_frequent_matched_caste_count = matched_caste_counts.max()
print(f"The most frequently appearing matched caste is: {most_frequent_matched_caste} with {most_frequent_matched_caste_count} occurrences.")

# Least frequent matched caste
least_frequent_matched_caste = matched_caste_counts.idxmin()
least_frequent_matched_caste_count = matched_caste_counts.min()
print(f"The least frequently appearing matched caste is: {least_frequent_matched_caste} with {least_frequent_matched_caste_count} occurrences.")

# Step 8: Caste occurrence count for each state
unique_castes_per_state = customer_data[customer_data['Caste Match'] == 1].groupby('State')['Matched Castes'].nunique()
print("\nUnique Castes Present in Each State:")
print(unique_castes_per_state)

# Assuming normalized_caste_names is already defined
# and contains the extracted caste names in lowercase

# Create a mask for 'General' caste customers
general_mask = customer_data['Caste'].str.lower() == 'general'

# Get the Customer IDs of 'General' caste customers
general_customer_data = customer_data.loc[general_mask]

# Initialize a set to store distinct Customer IDs that match ST castes
matching_customer_ids = set()

# Loop through each general customer
for _, row in general_customer_data.iterrows():
    # Get all relevant names and normalize them
    names_to_check = [
        row['Full Name'],
        row['Father Name'],
        row['Mother Name'],
        row['Spouse Name']
    ]

    # Normalize names and filter out NaN values
    normalized_names = [name.lower() for name in names_to_check if pd.notna(name)]

    # Check if any normalized name matches the ST caste names
    if any(name in normalized_caste_names for name in normalized_names):
        matching_customer_ids.add(row['Customer ID'])

# Count the distinct Customer IDs that matched
distinct_general_customer_ids_count = len(matching_customer_ids)

# Print the count of distinct 'General' caste customers matched as ST
print(f"Number of 'General' caste customers matched as ST: {distinct_general_customer_ids_count}")

# Filter and display first 5 entries with Customer IDs marked as General but matched with ST
entries_to_check = customer_data[customer_data['Customer ID'].isin(distinct_general_customer_ids)].head(5)
print("\nEntries with Customer IDs marked as General but matched as ST:")
print(entries_to_check)

# Step 10: Visualization for most and least frequent matched castes
plt.figure(figsize=(12, 6))
sns.barplot(x=matched_caste_counts.index[:10], y=matched_caste_counts.values[:10], palette='coolwarm')
plt.title('Top 10 Most Frequently Matched Castes')
plt.xlabel('Caste Names')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=matched_caste_counts.index[-10:], y=matched_caste_counts.values[-10:], palette='viridis')
plt.title('Top 10 Least Frequently Matched Castes')
plt.xlabel('Caste Names')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Step 11: Distinct loan ID count by state for matched castes
statewise_loan_counts = customer_data[customer_data['Caste Match'] == 1].groupby('State')['Loan ID'].nunique().reset_index()
statewise_loan_counts.columns = ['State', 'Distinct Loan ID Count']

formatted_table = tabulate(statewise_loan_counts, headers='keys', tablefmt='fancy_grid')
print("\n**************************************************************")
print("               State-wise Distinct Loan ID Count")
print("**************************************************************\n")
print(formatted_table)
print("\n**************************************************************\n")

# Step 12: Most frequently matched names
matched_name_counts = pd.Series(matched_names).value_counts()
most_frequent_matched_name = matched_name_counts.idxmax()
most_frequent_matched_name_count = matched_name_counts.max()
print(f"The most frequently appearing matched name is: {most_frequent_matched_name} with {most_frequent_matched_name_count} occurrences.")

# Visualization for most frequent matched names
plt.figure(figsize=(12, 6))
sns.barplot(x=matched_name_counts.index[:10], y=matched_name_counts.values[:10], palette='magma')
plt.title('Top 10 Most Frequently Matched Names')
plt.xlabel('Names')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Step 13: Additional visualizations
# Plot state-wise match percentages
plt.figure(figsize=(12, 6))
match_percentage_by_state.plot(kind='bar', color='skyblue')
plt.title('State-wise Match Percentage')
plt.xlabel('State')
plt.ylabel('Match Percentage (%)')
plt.xticks(rotation=45)
plt.show()

# Plot distinct caste count by state
plt.figure(figsize=(12, 6))
unique_castes_per_state.plot(kind='bar', color='orange')
plt.title('Unique Castes Present in Each State')
plt.xlabel('State')
plt.ylabel('Number of Unique Castes')
plt.xticks(rotation=45)
plt.show()

# Step 14: Customer count by caste type (General vs. others)
caste_type_counts = customer_data['Caste'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=caste_type_counts.index, y=caste_type_counts.values, palette='Set2')
plt.title('Customer Count by Caste Type')
plt.xlabel('Caste Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```
